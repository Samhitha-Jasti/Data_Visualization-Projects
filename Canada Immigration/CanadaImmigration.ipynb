```python
#Sorting the 'Total' column to get the top 5 countries using pandas sort_values() method.
##inplace = True paramemter saves the changes to the original df_can dataframe
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)

# passing the top 5 entries
df_top5 = df.head(5)
df_top5
```

```python
df_bottom5=df.tail(5)
df_bottom5=df_bottom5[years].transpose()
df_bottom5
```

```python
# transposing the dataframe
df_top5 = df_top5[years].transpose() 
df_top5
```

```python
#Plotting Area graph of top 5 countries
# Changing index values of df_top5 to type integer for plotting
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Top 5 immigration countries to Canada')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
```

```python
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Top 5 immigration countries to Canada')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
```
